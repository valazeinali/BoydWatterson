```python
# OAS Outlier Adjustment
q1 = df["OAS"].quantile(0.25)
q3 = df["OAS"].quantile(0.75)

iqr = q3 - q1

lower_bound = q1 -(1.5 * iqr) 
upper_bound = q3 +(1.5 * iqr)

print("Lower Bound Fence: " + str(lower_bound))
print("Upper Bound Fence: " + str(upper_bound))

for x in df["OAS"]:
    if x > upper_bound:
         x = upper_bound
    df.replace(to_replace = x, 
                 value =x)
    if x < lower_bound:
         x = lower_bound
    df.replace(to_replace = x, 
                 value =x)
        
    print(x)
```
