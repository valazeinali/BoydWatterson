```python
def lower_bound(df):
    q1 = df.quantile(0.25)
    q3 = df.quantile(0.75)

    iqr = q3 - q1

    lower_bounds = q1 -(1.5 * iqr) 
    
    return lower_bounds
    
```

```python
def upper_bound(df):
    q1 = df.quantile(0.25)
    q3 = df.quantile(0.75)

    iqr = q3 - q1

    upper_bounds = q3 +(1.5 * iqr) 
    
    return upper_bounds
    
```
