```python
def lower_bound(df):
    q1 = df.quantile(0.25)
    q3 = df.quantile(0.75)

    iqr = q3 - q1

    lower_bounds = q1 -(1.5 * iqr) 
    
    return lower_bounds
    
```

```python
def upper_bound(df):
    q1 = df.quantile(0.25)
    q3 = df.quantile(0.75)

    iqr = q3 - q1

    upper_bounds = q3 +(1.5 * iqr) 
    
    return upper_bounds
    
```

# Create Excess returns

```python
# Creating Excess Returns 3,6,9,12,24,36 for AGG
df_3_Agg = df1.rolling(3,win_type='boxcar',center=True).sum().shift(-1)
df_6_Agg = df1.rolling(6,win_type='boxcar',center=True).sum().shift(-3)
df_9_Agg = df1.rolling(9,win_type='boxcar',center=True).sum().shift(-4)
df_12_Agg = df1.rolling(12,win_type='boxcar',center=True).sum().shift(-6)
df_24_Agg = df1.rolling(24,win_type='boxcar',center=True).sum().shift(-12)
df_36_Agg = df1.rolling(36,win_type='boxcar',center=True).sum().shift(-18)
```
