```python
import warnings
warnings.filterwarnings("ignore")
```

```python
df_ABS = pd.read_excel('Hyp_3_matrix.xlsx', wb = "ABS") 
df_MBS = pd.read_excel('Hyp_3_matrix.xlsx', wb = "MBS")
df_CMBS = pd.read_excel('Hyp_3_matrix.xlsx', wb = "CMBS") 
df_HY = pd.read_excel('Hyp_3_matrix.xlsx', wb = "HY")
df_IG = pd.read_excel('Hyp_3_matrix.xlsx', wb = "IG")
df_FRPS = pd.read_excel('Hyp_3_matrix.xlsx', wb = "FRPS")
df_EM = pd.read_excel('Hyp_3_matrix.xlsx', wb = "EM")
```

```python
df_ABS.head(10)
```

```python
# Selecting all rows where we are 75th percentile and benchmark is tres.
df_ABS_75_tres = df_ABS[(df_ABS["OAS Percentile"] == 85)& (df_ABS['Benchmark'] == 'tres')]
df_ABS_75_tres.head(6)
```

```python
fig = make_subplots(specs=[[{"secondary_y": True}]])

Month = df_ABS_75_tres["Month"]
X_Percentile_Mean = df_ABS_75_tres["OAS Percentile Mean Excess Return"]
Population_Mean = df_ABS_75_tres["Population Mean Excess Return"]
p = df_ABS_75_tres["P-Value"]

fig.add_trace(go.Scatter(x=Month, y=X_Percentile_Mean, name='75th_Percentile_Mean'),
                         secondary_y=False,)
fig.add_trace(go.Scatter(x=Month, y=Population_Mean, name='Population_Mean'),
                         secondary_y=False,)
fig.add_trace(go.Scatter(x=Month, y=p, name='p-value',),
                         secondary_y=True,)

# Add figure title
fig.update_layout(
    title_text='Analysis of 75th percentile OAS Excess Returns, '+str(df_ABS_75_tres.iloc[1, 0])+' to tres.'
    ,width=1000, height=800
)

# Set x-axis title
fig.update_xaxes(title_text="Month Out")

fig.update_yaxes(title_text="<b>Mean</b> ", secondary_y=False)
fig.update_yaxes(title_text="<b>P-Value</b> ", secondary_y=True)

fig.show()
```
