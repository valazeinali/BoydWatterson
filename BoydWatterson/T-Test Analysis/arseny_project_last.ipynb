# Drop Outliers!

```python
df = df.drop(columns = ['Date'])

q1 = df.quantile(0.25)
q3 = df.quantile(0.75)

iqr = q3 - q1

lower_bound = q1 -(1.5 * iqr) 
upper_bound = q3 +(1.5 * iqr)

print(lower_bound)
print(upper_bound)

df = df[(df > lower_bound) & (df < upper_bound)] # gets rid of outliers on Agg Excess Returns
df['Date'] = df_Date
df.tail()
```

```python
df1 = df.drop(columns = ['Date'])

q1 = df1.quantile(0.25)
q3 = df1.quantile(0.75)

iqr = q3 - q1

lower_bound = q1 -(1.5 * iqr) 
upper_bound = q3 +(1.5 * iqr)

print(lower_bound)
print(upper_bound)

df1 = df1[(df1 > lower_bound) & (df1 < upper_bound)] # gets rid of outliers on Agg Excess Returns
df1['Date'] = df1_Date
df1.head()
```

```python
df_OAS = df_OAS.drop(columns = ['Date'])

q1 = df_OAS.quantile(0.25)
q3 = df_OAS.quantile(0.75)

iqr = q3 - q1

lower_bound = q1 -(1.5 * iqr) 
upper_bound = q3 +(1.5 * iqr)

print(lower_bound)
print(upper_bound)

df_OAS = df_OAS[(df_OAS > lower_bound) & (df_OAS < upper_bound)] # gets rid of outliers on Agg Excess Returns
df_OAS['Date'] = df_OAS_Date
df_OAS.head(20)
```

# Create Excess returns

```python
# Creating Excess Returns 3,6,9,12,24,36 for AGG
df_3_Agg = df.rolling(3,win_type='boxcar',center=True).sum().shift(-1)
df_3_Agg['Date'] = df_Date
df_6_Agg = df.rolling(6,win_type='boxcar',center=True).sum().shift(-3)
df_6_Agg['Date'] = df_Date
df_9_Agg = df.rolling(9,win_type='boxcar',center=True).sum().shift(-4)
df_9_Agg['Date'] = df_Date
df_12_Agg = df.rolling(12,win_type='boxcar',center=True).sum().shift(-6)
df_12_Agg['Date'] = df_Date
df_24_Agg = df.rolling(24,win_type='boxcar',center=True).sum().shift(-12)
df_24_Agg['Date'] = df_Date
df_36_Agg = df.rolling(36,win_type='boxcar',center=True).sum().shift(-18)
df_36_Agg['Date'] = df_Date
```

```python
# Joining the datasets
pd.options.display.max_columns  = 150 # want to see all columns

df_Agg_3 = pd.merge(df_3_Agg, df_OAS, on='Date', how='left')
df_Agg_6 = pd.merge(df_6_Agg, df_OAS, on='Date', how='left')
df_Agg_9 = pd.merge(df_9_Agg, df_OAS, on='Date', how='left')
df_Agg_12 = pd.merge(df_12_Agg, df_OAS, on='Date', how='left')
df_Agg_24 = pd.merge(df_24_Agg, df_OAS, on='Date', how='left')
df_Agg_36 = pd.merge(df_36_Agg, df_OAS, on='Date', how='left')
```
