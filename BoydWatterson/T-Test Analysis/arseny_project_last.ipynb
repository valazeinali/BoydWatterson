```python
# Creating Excess Returns 3,6,9,12,24,36 for AGG
df_3_Agg = df.rolling(3,win_type='boxcar',center=True).sum().shift(-1)
df_6_Agg = df.rolling(6,win_type='boxcar',center=True).sum().shift(-3)
df_9_Agg = df.rolling(9,win_type='boxcar',center=True).sum().shift(-4)
df_12_Agg = df.rolling(12,win_type='boxcar',center=True).sum().shift(-6)
df_24_Agg = df.rolling(24,win_type='boxcar',center=True).sum().shift(-12)
df_36_Agg = df.rolling(36,win_type='boxcar',center=True).sum().shift(-18)
```

```python
df_3_Agg['Date'] = df['Date']
df_6_Agg['Date'] = df['Date']
df_9_Agg['Date'] = df['Date']
df_12_Agg['Date'] = df['Date']
df_24_Agg['Date'] = df['Date']
df_36_Agg['Date'] = df['Date']
```

```python
df_OAS = df_OAS.dropna()
df_OAS.head()
```

```python
# Joining the datasets
pd.options.display.max_columns  = 150 # want to see all columns

df_Agg_3 = pd.merge(df_3_Agg, df_OAS, on='Date', how='right')
df_Agg_6 = pd.merge(df_6_Agg, df_OAS, on='Date', how='right')
df_Agg_9 = pd.merge(df_9_Agg, df_OAS, on='Date', how='right')
df_Agg_12 = pd.merge(df_12_Agg, df_OAS, on='Date', how='right')
df_Agg_24 = pd.merge(df_24_Agg, df_OAS, on='Date', how='right')
df_Agg_36 = pd.merge(df_36_Agg, df_OAS, on='Date', how='right')
```
