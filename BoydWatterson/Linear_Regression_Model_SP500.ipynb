# Lets do the same analysis, but with S&P 500 returns

```python
import numpy as np
import pandas as pd
import seaborn as sns
import math
sns.set_palette('husl')
import matplotlib.pyplot as plt
%matplotlib inline
%pylab inline
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
data = pd.read_csv('Managers.csv')
```

```python
data1 = pd.read_csv('SP500.csv')
```

```python
data.head()
```

```python
data1.head()
```

## Right Outer Join on the dates

```python
newData = pd.merge(data, data1, on='Date', how='right') 
```

```python
newData.head()
```

## Linear Regression model 

```python
fig=plt.figure(figsize=(15, 45)) # Manager 
sns.lmplot('SP_500_Return', 'Manager_A_Monthly_Return_Percent', data = newData, height=8, fit_reg=True)
plt.title('Relationship between Manager A and S&P 500', size=24)
```

```python
fig=plt.figure(figsize=(15, 45)) # Manager 
sns.lmplot('SP_500_Return', 'Manager_B_Monthly_Return_Percent', data = newData, height=8, fit_reg=True)
plt.title('Relationship between Manager B and S&P 500', size=24)
```

```python
fig=plt.figure(figsize=(15, 45)) # Manager 
sns.lmplot('SP_500_Return', 'Manager_C_Monthly_Return_Percent', data = newData, height=8, fit_reg=True)
plt.title('Relationship between Manager C and S&P 500', size=24)
```

# Correlation Matrix to show which managers have most correlated return amongst the S&P 500.


```python
#Pairwise correlation between attributes
fig=plt.figure(figsize=(10, 10))
sns.heatmap(newData[['SP_500_Return', 'Manager_A_Monthly_Return_Percent', 'Manager_B_Monthly_Return_Percent', 'Manager_C_Monthly_Return_Percent' ]].corr(), vmin = -1, vmax=1, annot=True)
plt.show()
```
